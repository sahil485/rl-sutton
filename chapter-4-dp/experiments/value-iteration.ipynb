```python
import numpy as np
```

```python
def init_grid(grid: np.ndarray):
    counter = 1
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if (i == 0 and j == 0) or (i == 3 and j == 3):
                grid[i, j] = 0
            else:
                grid[i, j] = counter
                counter += 1               
```

```python
from typing import Tuple

def get_next_state(state: Tuple[int, int], action: str) -> Tuple[int, int]:
    i, j = state
    
    if action == 'up':
        next_i, next_j = max(0, i - 1), j
    elif action == 'down':
        next_i, next_j = min(3, i + 1), j
    elif action == 'left':
        next_i, next_j = i, max(0, j - 1)
    elif action == 'right':
        next_i, next_j = i, min(3, j + 1)
    
    return (next_i, next_j)
```

```python
def init_state_value_func(): 
    return np.zeros((4, 4))
```

```python
def init_policy():
    return np.ones((4, 4, 4)) * 0.25
```

```python
from typing import List

def action_to_index(action: str, actions: List[str] = ['up', 'down', 'left', 'right']):
    return actions.index(action)
```

```python
from typing import List

def index_to_action(index: int, actions: List[str] = ['up', 'down', 'left', 'right']):
    return actions[index]
```

```python
from typing import Tuple, Dict, List

def calculate_p(current_state: Tuple[int], action, next_state: Tuple[int], reward: int, adjacency: Dict[Tuple[int, int], List[Tuple[int, int]]]):
    if next_state in adjacency[current_state] and reward == -1:
        return 1
    return 0
```

```python
THETA = 1e-10
DISCOUNT_FACTOR = 1

v = init_state_value_func()

grid = np.zeros((4, 4))
init_grid(grid)

pi_a_given_s = 1/4 # pi(a | s)

actions = ['up', 'down', 'left', 'right']

policy = init_policy()
```

```python
import numpy as np

def run_value_iteration(grid, actions, discount_factor, theta):
    error = float('inf')

    v = init_state_value_func()
    policy = init_policy()
    iterations = 0

    while error > theta:
        v_new = v.copy()
        for state in [(i, j) for i in range (len(grid)) for j in range(len(grid[0]))]:
            if state == (0, 0) or state == (3, 3):
                continue
                
            best_action = None
            best_value = float('-inf')
            for action in actions:
                next_state = get_next_state(state, action)
                reward = -1
                action_value = reward + discount_factor * v[next_state]
                
                if action_value > best_value:
                    best_value = action_value
                    best_action = action

            v_new[state] = best_value
            policy[state[0], state[1], :] = 0
            policy[state[0], state[1], actions.index(best_action)] = 1
            
        error = np.max(np.abs(v_new - v))
        v = v_new
        iterations += 1
    
    print(f"Value iteration converged in {iterations} iterations")
    
    return v, policy
```

```python
THETA = 1e-10
DISCOUNT_FACTOR = 0.9

v = init_state_value_func()

grid = np.zeros((4, 4))
init_grid(grid)

actions = ['up', 'down', 'left', 'right']

policy = init_policy()

run_value_iteration(grid, actions, discount_factor = DISCOUNT_FACTOR, theta = THETA)
```
